```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
x = 2*np.random.rand(100,1)
y= 4 + 3*x + np.random.randn(100,1)

m = 100
w=0
b=0

alpha = 0.01
cost = []
for i in range(500):
    y_pred = w*x +b
    summation_w=0
    summation_b=0
    cst = (1/2*m)*np.sum((y_pred-y)**2)
    cost.append(cst)
    for i in range(m):
        summation_w += (w*x[i]+b-y[i])*x[i]
        summation_b += (w*x[i]+b-y[i])
    
    dw = (1/m)*summation_w
    db = (1/m)*summation_b
    w = w-alpha*dw
    b = b-alpha*db

print(f"final value of the slope {w}")
print(f"final value of the interceptor {b}")

plt.scatter(x,y,color="blue",label="training data")
plt.plot(x,w*x+b,color="red",label="predicted line")
plt.xlabel('x')
plt.ylabel('y')
plt.title("linear regression of univariate")
plt.legend()
plt.grid(True)
plt.show()
plt.plot(cost)
plt.show()
```
